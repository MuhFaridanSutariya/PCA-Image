```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA, NMF
```

## import dataset

```python
df = pd.read_csv("data/lfw_people.csv")
df.head()
```

## Dataset splitting

```python
X = df.drop(columns="name").values # data image menggunakan numpy .values
y = df['name'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

## Visualize data

```python
idx = 1000
plt.imshow(X_train[idx].reshape(87, 65), cmap='gray')
plt.title(y_train[idx])
plt.axis('off')
```

## Find n-components

```python
pca = PCA().fit(X_train)
plt.plot(pca.explained_variance_ratio_.cumsum())
plt.xlabel("n-components")
plt.ylabel("Cumulative Explained Variance")
```

## Feature Extraction menggunakan Dimensional Reduction

```python
pca = PCA(300, whiten=True)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)
```

```python
print(X_train.shape) # hidden layer
print(X_train_pca.shape) # weight / code
print(pca.components_.shape) # feature
```

## Visualize components(feature)

```python
plt.figure(figsize=(15, 8))
for i in range(10):
    plt.subplot(2, 5, i+1)
    plt.imshow(pca.components_[i].reshape(87, 65), cmap="gray")
    plt.title(f"pca components ke-{i+1}")
    plt.axis("off");
```

## Reconstruct Image

```python
X_train_inv = pca.inverse_transform(X_train_pca)
```

```python
plt.figure(figsize=(8, 5))

plt.subplot(121)
plt.imshow(X_train[2].reshape(87, 65), cmap="gray")
plt.title(y_train[2])
plt.axis("off");

plt.subplot(122)
plt.imshow(X_train_inv[2].reshape(87, 65), cmap="gray")
plt.title(y_train[2])
plt.axis("off");
```
